**Description of the data**

In this dataset we can see the daily count of COVID 19 cases, hospitalizations and deaths in

Important variables:

York City. Is a numeric variable, discreet.

*'DEATH_COUNT': display the daily count of deaths due to covid 19 in New York City. Is a
numeric variable, discreet.

*'DATE_OF_INTEREST': display the date in which the count was taken in New York City. Is a
string variable, ordinal.

```python
def custom_cls(x):
    if (0 <= x <= 120):
        return '<= 120'
    elif (121 <= x <= 30):
        return '121 - 230'
    elif (231 <= x <= 500):
        return '231 - 500'
    else:
        return '> 500'
```

```python
df['case_counts'] = df.apply(lambda tbl: custom_cls(tbl['CASE_COUNT']), axis = 1)
df.head()
```

```python
case_date = df.groupby(['case_counts','CASE_COUNT']).size()
case_date = case_date.to_frame()
```

```python
case_date = pd.pivot_table(index =['case_counts','CASE_COUNT'],
                         aggfunc = len,
                         values = 'DEATH_COUNT',
                         data = df,
                         margins = True)
case_date 
```

```python
case_date = case_date.rename(columns = {0: 'Count'})
case_date = case_date.reset_index()
case_date.columns
```

```python
x = case_date.iloc[:,-1].sum()
x
case_date['%'] = (case_date.iloc[:,-1]/x)*100
case_date
```

```python
plot_order = ['<= 120', '121 - 230', '231 - 500', '> 500']
```

```python
fig = px.sunburst(case_date,
                values = 'DEATH_COUNT',
                path = ['DEATH_COUNT','CASE_COUNT'],
                color = 'case_counts',
                title = 'Number of respondents by AgeGroup and Gender'
                )
plt.savefig('sample_data/age_groupSun.png')
fig.show()
```
