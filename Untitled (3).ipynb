```python
import pandas as pd
import numpy as np
%ls datasets
```

```python
df = pd.read_csv('datasets/antropometria-dataset-2.csv')
```

```python
df.head(5)
```

```python
pd.pivot_table(df,
               index = ['desc_ent'],
               values = 'peso',
               columns = 'sexo',
               margins = True
              )
```

```python
pd.pivot_table(df,
               index = ['desc_ent'],
               values = 'peso',
               columns = 'sexo',
               aggfunc = ['median'],
               margins = True
              )
```

```python
df.groupby(['sexo', 'peso']).size()
```

```python
pd.pivot_table(df,
               index = ['desc_ent'],
               aggfunc = ['mean'],
               values = ['peso'],
               columns = 'sexo',
               margins = True
              )
```
