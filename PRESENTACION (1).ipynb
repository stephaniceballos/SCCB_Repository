# Description of the data

In this dataset we can see the daily count of COVID 19 cases, hospitalizations and deaths in Oceania. In the variables we can observe the day, death count, hospitalizations and cases, there is also a probable number of the same issues, this is because there may be cases that were not counted due to lack of information, etc. There are also variables in which we can observe the same theme but obtained weekly.

All variables are numeric, discrete, only the date is a string, ordinal.
