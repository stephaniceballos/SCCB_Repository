```python
df.describe()
```

```python
x = df.columns.tolist()
x.sort()
x
```

```python
pd.pivot_table(df,
               index = ['STNAME','CTYNAME'],
               aggfunc = ['median', 'mean'],
               values = ['BIRTHS2010', 'BIRTHS2011'],
               margins = True,
               columns = 'SUMLEV'
              )
```
