```python
sns.swarmplot(x = 'Gender',
            y = 'CWDistance',
            data = df
            )
```

## Box plot of Male data

```python
sns.boxplot(data = df.loc[df['Gender'] == 'M',
                          ['Age','Height', 'Wingspan', 'CWDistance','Score']
                         ]
           )
plt.xticks(rotation = 45, fontsize = 17)
plt.yticks(fontsize = 17)
plt.title('Box plot representation of the entire data', fontsize = 15)
plt.show()
```

### 1. Make a comparison of 'Scores' of 'Male' and 'Female' data

```python
fig = plt.figure()
ax1 = fig.add_subplot(1,2,1)

sns.boxplot(data = df.loc[df['Gender']=='M',
                          ['Score']
                         ]
           )

ax2 = fig.add_subplot(1,2,2)

sns.boxplot(data = df.loc[df['Gender']=='F',
                          ['Score']
                         ]
           )
ax1.title.set_text('Male box plot')
ax2.title.set_text('Female box plot')
plt.show()
```
