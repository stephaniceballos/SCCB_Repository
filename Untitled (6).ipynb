```python
df = pd.read_excel('datasets/Canada.xlsx',
                   engine = 'openpyxl',
                   sheet_name = 'Canada by Citizenship',
                   skiprows = range(20),
                   skipfooter = 2
                  )
#df.head()
```

```python
df.columns
```

```python
df.drop([ 'Type','Coverage', 'AREA', 'REG', 'DEV','Unnamed: 43', 'Unnamed: 44',
       'Unnamed: 45', 'Unnamed: 46', 'Unnamed: 47', 'Unnamed: 48',
       'Unnamed: 49', 'Unnamed: 50',],
        axis = 1,
        inplace = True
       )
#df.columns
```

```python
df.columns
df.rename(columns = {'OdName': 'Country', 'AreaName': 'Continent', 'RegName': 'Region'},
         inplace = True
         )
```

```python
df.columns
```

```python
df.columns = list(map(str, df.columns))
df.columns
```

```python
df.set_index('Country', inplace = True)
df.head()
```

```python
df['Total'] = df.iloc[:,
                      2:].sum(axis = 1)
df.head(2)
```

```python
df.iloc[:,3]
```

```python
j_data = df.loc['Japan','1980':'2013']
j_data.plot()
```

```python
ic_data = df.loc[['India','China'],'1998':'2013']
```

```python
nic = ic_data.transpose()
```

```python
nic.plot(figsize = (20,10))
plt.xlabel('Years', fontsize = 17)
plt.xticks(fontsize = 17, rotation = 45)
plt.ylabel('Number of imigrants', fontsize = 17)
plt.yticks(fontsize = 17)
plt.title('Imigrantion to Canada from India and China during 1980 to 2013', fontsize = 17)
```
